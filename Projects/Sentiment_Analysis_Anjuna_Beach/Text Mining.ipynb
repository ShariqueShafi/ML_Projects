```python
# Prepare data for Sankey diagram. The current year_sentiment data is already in a suitable format
# if we consider Year as a 'source' and Sentiment as a 'target' for flow.
# However, Sankey diagrams typically visualize flow *between* distinct nodes.
# For sentiment analysis over years, we can visualize the 'flow' of reviews from years to sentiment categories.

# Let's create a DataFrame suitable for Sankey, where 'source' could be the Year and 'target' the Sentiment.
# Alternatively, we can show the distribution of sentiment within each year.

# Option 1: Visualize 'flow' from Year to Sentiment
# This means each year acts as a source node and each sentiment acts as a target node.
# The 'links' data should have 'source', 'target', and 'value' columns.

# Create a unique identifier for each year and sentiment combination
links_sankey = year_sentiment.copy()
links_sankey['source'] = links_sankey['Year'].astype(str)
links_sankey['target'] = links_sankey['Sentiment']
links_sankey['value'] = links_sankey['Count']

# In Holoviews Sankey, nodes are implicitly created from source/target strings.
# Let's visualize this.

hv.Sankey(links_sankey, kdims=['source', 'target'], vdims=['value']).opts(
    opts.Sankey(height=600, width=900, label_position='right', edge_color='target', edge_cmap=['yellow', 'green', 'red'])
)
```

```python
import nltk ###Text Preprocessing
from nltk.corpus import stopwords ###Text Preprocessing stopwords
from nltk.stem import WordNetLemmatizer ### Text Preprocessing Lemmatizer
from gensim.models import LdaModel ### LDA Model
from gensim import corpora ### LDA Corpora
from gensim.models.coherencemodel import CoherenceModel
nltk.download('stopwords') ### Downloading Stopwords
nltk.download('wordnet')
```

```python
## Lower Case ##
df['Text'] = df['Text'].str.lower()
```

```python
## Remove symbol and Number ##
df['Text']=df['Text'].str.replace('[^a-z\s]',"",regex=True)
```

## REMOVE STOPWORDS

```python
domain_stopwords = {
    "good", "nice", "best", "one", "place", "visit", "time",
    "like", "really", "much", "many", "great", "beautiful",
    "goa", "anjuna", "beach"
}
```

```python
stop_words = set(stopwords.words('english')).union(domain_stopwords)
Lemmatizer = WordNetLemmatizer()

def clean_text(text):
  return[
      Lemmatizer.lemmatize(word)
      for word in text.split()
      if word not in stop_words and len(word)>2
  ]
df['tokens'] = df['Text'].apply(clean_text)
```

```python
df = df[df['tokens'].str.len()>=4]
```

```python
dictionary = corpora.Dictionary(df['tokens'])
dictionary.filter_extremes(no_below=5, no_above=0.3)
corpus = [dictionary.doc2bow(text) for text in df['tokens']]
```

```python
num_topics = 3
lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    random_state=42,
    passes=30,
    alpha="asymmetric",
    eta="auto"
)
```

```python
print("\nLDA Topics:\n")
for i, topic in lda_model.print_topics(num_topics=num_topics, num_words=10):
    print(f"Topic {i}: {topic}")
```

```python
coherencemodel = CoherenceModel(model=lda_model, texts=df['tokens'], dictionary=dictionary, coherence='c_v')
coherence_score = coherencemodel.get_coherence()
print(f"Coherence Score: {coherence_score}")  ## <0.3 = poor, >0.6 = excelent ##
```
